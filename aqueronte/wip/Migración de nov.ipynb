```python
def get_sdict_items(path):
    import json
    
    with open(path) as f:
        json_data = json.load(f, encoding='utf-8')
    #HARDCODE: PARA TRAINING
    chicho = json_data['phraselist'].items()
    chicho_out = {}
    
    for ent, val in chicho:
        value = val['words']
        chicho_out[ent] = value
    
    with open('sdict_items.json', 'w') as f:
        output_js = json.dump(chicho_out, f, ensure_ascii=False, indent=4)
    
    return output_js
```

```python
#Obtengo dominios posibles

posible_domains = []
domains = []

for intent, utt in json_data['intents'].items():
    #chicho = intent.replace('intent.', '')
    posible_domains.append(intent)

#print(posible_domains)

#Obtengo los dominos
posdom = re.compile(r'(intent\.[a-z]+)\.')

for dom in posible_domains:
    a = re.findall(posdom, dom)
    domains.extend(a)
    domains = list(set(domains))

domains
```

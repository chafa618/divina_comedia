```python
def load_json(path):
    with open(path) as f:
        archivo = json.load(f)
    return archivo
```

```python
def get_key(archivo):
    entidadRe = re.compile(r'\[.+?\]')
    entidades = []
    dic = {}

    for intent in archivo['intents'].keys():
        dic[intent] = []

        for utterance in archivo['intents'][intent]:
            ent = re.findall(entidadRe,utterance)

            if len(ent) >= 1:  
                dic[intent].append(ent)
    return dic 
```

```python
def count_ents_for_intent(dic):
    claves = get_key(dic)
    
    output = {}
    for intent in claves:
        output[intent] = {}
        
        for utt in claves[intent]:
            #print(utt)
            key = ",".join(claves[intent][claves[intent].index(utt)])
            output[intent][key] = claves[intent].count(utt) 
    
    with open('output.json', 'w') as file:
        json.dump(output, file, indent=4)
    
    return output
```
